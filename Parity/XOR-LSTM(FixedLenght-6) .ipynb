```python
import tensorflow as tf
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
def generateData(SeqCount=100000):
    train_y = np.zeros((SeqCount, 2))
    seqs = np.random.choice(2, [SeqCount, 6])
    for i,seq in enumerate(seqs):
        train_y[i][sum(seq)%2] = 1
    return seqs.reshape((SeqCount, 6, 1)), train_y
```

```python
train_x, train_y = generateData(SeqCount=100000)
```

```python
model = Sequential() 
batch_size, timesteps, data_dim = None, None, 1
model.add(LSTM(10, return_sequences=False, stateful=False,
               batch_input_shape=(batch_size, timesteps, data_dim)))
model.add(Dense(2, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(train_x, train_y, batch_size=64, epochs=100)
```
